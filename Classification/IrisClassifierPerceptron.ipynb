```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import math
```

```python
from sklearn.datasets import load_iris
iris=load_iris()
```

```python
iris_df=pd.DataFrame(iris.data, columns=iris.feature_names)
iris_df['species']=[iris.target_names[i] for i in iris.target]
iris_df.head()
```

```python
y=iris_df.iloc[0:100,4].values
y=np.where(y=='setosa',-1,1)
#print(y)

X=iris_df.iloc[0:100,[0,1]].values #interesting case [0,1]
#print(X)

plt.scatter(X[:50,0],X[:50,1], color='red', marker='o')
plt.scatter(X[50:100,0],X[50:100,1], color='blue', marker='x')
plt.legend()
```

```python
H=np.zeros([100,3])
for i in range(50):
  H[i,:]=[X[i,0], X[i,1], -1]
  pass

for i in range(50,100):
  H[i,:]=[-X[i,0], -X[i,1], 1]
  pass

#print(H)
```

```python
n=100;
k=0;
S=1000 #epochs number
K=0
k1=0

Theta=np.zeros([3,1])
i=0
s=0
while i<=n*S:
  if math.floor(i/n)==s+1:
    plt.scatter(i/n,k,color='black')
    k=0
    s=s+1
  #print('i=',i)
  j=i-(n)*math.floor(i/((n)))
  #print(j)
  if np.dot(H[j,:],Theta)>0:
    Theta=Theta
  else:
    a=H[j,:].reshape(3,-1)
    Theta=Theta+a
    k=k+1
    k1=k1+1
    print(i)
    print('epoch=',math.floor(i/n)+1)
    #plt.scatter(i/100,k,color='black')
  i=i+1

print(k1)
plt.xlabel('epochs')
```

```python
X1=np.linspace(min(X[:,0]),max(X[:,0]),1000)
Y1=1/Theta[1]*(Theta[2]-Theta[0]*X1)

plt.plot(X1,Y1,color='black')
plt.scatter(X[:50,0],X[:50,1], color='red', marker='o')
plt.scatter(X[50:100,0],X[50:100,1], color='blue', marker='x')
```

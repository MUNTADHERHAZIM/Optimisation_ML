```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
```

```python
Sonar_dataset = pd.read_csv('Sonar.csv')
Sonar_dataset.head()
```

```python
x=Sonar_dataset[['V1']].values
n=x.shape[0]
print(n)

Y=Sonar_dataset[['Class']].values
print(Y.T)
XX=Sonar_dataset[0:n].values

X=np.zeros([n,60])
for i in range(n):
  for j in range(60):
    X[i,j]=XX[i,j]

```

```python
knn=KNeighborsClassifier(n_neighbors=1, p=2, metric='minkowski')
knn._fit(X,Y.ravel())
yy=X[0:3,:]
xx=np.zeros([1,60])
for i in range(60):
  #print(i)
  xx[0,i]=0.1**i
print(yy.shape)
print(knn.predict(X[112:113,:]))
```

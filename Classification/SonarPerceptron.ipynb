```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math
```

```python
Sonar_dataset = pd.read_csv('Sonar.csv')
Sonar_dataset.head()
```

```python
x=Sonar_dataset[['V1']].values
n=x.shape[0]
print(n)

Y=Sonar_dataset[['Class']].values

XX=Sonar_dataset[0:n].values

X1=np.zeros([n,60])
for i in range(n):
  for j in range(60):
    X1[i,j]=XX[i,j]

X=np.zeros([n,61])
for i in range(n):
  X[i,60]=-1
  for j in range(60):
    X[i,j]=X1[i,j]


```

```python
H=np.zeros([n,61])
for i in range(n):
  if Y[i]==1:
    H[i,:]=X[i,:]
  else:
    H[i,:]=-X[i,:]
```

```python
k=0;
S=10000
K=0
k1=0

Theta=np.zeros([61,1])
i=0
s=0
while i<=n*S:
  if math.floor(i/n)==s+1:
    plt.scatter(i/n,k,color='black')
    k=0
    s=s+1
  #print('i=',i)
  j=i-(n)*math.floor(i/((n)))
  #print(j)
  if np.dot(H[j,:],Theta)>0:
    Theta=Theta
  else:
    a=H[j,:].reshape(61,-1)
    Theta=Theta+a
    k=k+1
    k1=k1+1
    #print(i)
    print('epoch=',math.floor(i/n)+1)
    #plt.scatter(i/100,k,color='black')
  i=i+1

print(k1)
plt.xlabel('epochs')
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.cluster import DBSCAN
```

```python
from sklearn.datasets import make_moons

X,y=make_moons(n_samples=1000, noise=0.05, random_state=0)
plt.scatter(X[:,0],X[:,1])
#print(y)
```

```python
kmeans=KMeans(n_clusters=2)
y_km=kmeans.fit_predict(X)
print(y_km)
plt.scatter(X[y_km==0,0],X[y_km==0,1],c='black',label='cluster 1')
plt.scatter(X[y_km==1,0],X[y_km==1,1],c='red',label='cluster 2')
plt.legend()
```

```python
dbscan=DBSCAN(eps=0.2,min_samples=5,metric='euclidean')
y_db=dbscan.fit_predict(X)
dbscan.labels_
plt.scatter(X[y_db==0,0],X[y_db==0,1],c='black',label='cluster 1')
plt.scatter(X[y_db==1,0],X[y_db==1,1],c='red',label='cluster 2')
plt.legend()
```

```python
from sklearn.datasets import make_circles

X,y=make_circles(n_samples=5000, factor=0.5, noise=0.05)
plt.scatter(X[:,0],X[:,1])
```

```python
kmeans=KMeans(n_clusters=2)
y_km=kmeans.fit_predict(X)
print(y_km)
plt.scatter(X[y_km==0,0],X[y_km==0,1],c='black',label='cluster 1')
plt.scatter(X[y_km==1,0],X[y_km==1,1],c='red',label='cluster 2')
plt.legend()
```

```python
dbscan=DBSCAN(eps=0.1,min_samples=5,metric='euclidean')
y_db=dbscan.fit_predict(X)
dbscan.labels_
plt.scatter(X[y_db==0,0],X[y_db==0,1],c='black',label='cluster 1')
plt.scatter(X[y_db==1,0],X[y_db==1,1],c='red',label='cluster 2')
```

```python
from sklearn.datasets import make_blobs

X,y=make_blobs(n_samples=1000, centers=5, random_state=1)
plt.scatter(X[:,0],X[:,1])


```

```python
kmeans=KMeans(n_clusters=5)
y_km=kmeans.fit_predict(X)
print(y_km)
plt.scatter(X[y_km==0,0],X[y_km==0,1],c='black',label='cluster 1')
plt.scatter(X[y_km==1,0],X[y_km==1,1],c='red',label='cluster 2')
plt.scatter(X[y_km==2,0],X[y_km==2,1],c='green',label='cluster 3')
plt.scatter(X[y_km==3,0],X[y_km==3,1],c='blue',label='cluster 4')
plt.scatter(X[y_km==4,0],X[y_km==4,1],c='yellow',label='cluster 5')
plt.legend()
```

```python
dbscan=DBSCAN(eps=0.4,min_samples=5,metric='euclidean')
y_db=dbscan.fit_predict(X)
dbscan.labels_
plt.scatter(X[y_db==0,0],X[y_db==0,1],c='black',label='cluster 1')
plt.scatter(X[y_db==1,0],X[y_db==1,1],c='red',label='cluster 2')
plt.scatter(X[y_db==2,0],X[y_db==2,1],c='green',label='cluster 3')
plt.scatter(X[y_db==3,0],X[y_db==3,1],c='yellow',label='cluster 4')
plt.scatter(X[y_db==4,0],X[y_db==4,1],c='orange',label='cluster 4')
plt.scatter(X[y_db==-1,0],X[y_db==-1,1],c='blue',label='noisy')
plt.legend()
```

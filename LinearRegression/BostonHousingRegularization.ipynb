```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from mlxtend.plotting import scatterplotmatrix
from mpl_toolkits.mplot3d import Axes3D
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
from sklearn.linear_model import ElasticNet
```

```python
dataset = pd.read_csv('BostonHousing.csv')
dataset.head()
```

```python
X=dataset[['rm']].values
Y=dataset['medv'].values
n=X.shape[0]
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=1/10,random_state=0)

plt.scatter(X_train,Y_train,color='blue')
plt.scatter(X_test,Y_test,color='red')
```

```python
LMS = LinearRegression()
LMS.fit(X_train,Y_train)

rid1 = Ridge(alpha=50)
rid1.fit(X_train,Y_train)
rid2 = Ridge(alpha=200)
rid2.fit(X_train,Y_train)
```

```python
plt.scatter(X_train,Y_train,color='black')
plt.plot(X_train, LMS.predict(X_train), color='red')
plt.plot(X_train, rid1.predict(X_train), color='blue')
plt.plot(X_train, rid2.predict(X_train), color='green')
```

```python
lasso1 = Lasso(alpha=1)
lasso1.fit(X_train,Y_train)
lasso2 = Lasso(alpha=2)
lasso2.fit(X_train,Y_train)

plt.scatter(X_train,Y_train,color='black')
plt.plot(X_train, LMS.predict(X_train), color='red')
plt.plot(X_train, lasso1.predict(X_train), color='blue')
plt.plot(X_train, lasso2.predict(X_train), color='green')
```

```python
en1 = ElasticNet(alpha=1, l1_ratio=1.0)
en1.fit(X_train,Y_train)
en2 = ElasticNet(alpha=2, l1_ratio=0.5)
en2.fit(X_train,Y_train)

plt.scatter(X_train,Y_train,color='black')
plt.plot(X_train, LMS.predict(X_train), color='red')
plt.plot(X_train, en1.predict(X_train), color='blue')
plt.plot(X_train, en2.predict(X_train), color='green')
```

```python
XX=dataset[['rm','lstat','indus','nox','crim','b','ptratio','tax','rad','dis','age','chas','zn']].values
RidgeMult = Ridge(alpha=10)
RidgeMult.fit(XX,Y)
print(RidgeMult.coef_)
```

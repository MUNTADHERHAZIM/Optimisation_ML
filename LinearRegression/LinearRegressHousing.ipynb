```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from mlxtend.plotting import scatterplotmatrix
from mpl_toolkits.mplot3d import Axes3D
```

```python
dataset = pd.read_csv('BostonHousing.csv')
dataset.head()
```

Диаграммы рассеяния

```python
cols=['lstat','indus','nox','rm','medv']
scatterplotmatrix(dataset[cols].values, figsize=(10,10), names=cols)
plt.tight_layout() #для аккуратного включения подграфиков - подстраивает размер
```

Линейная связь мужду rm и medv - построим регрессионную модель

```python
X=dataset[['rm']].values
Y=dataset['medv'].values
n=X.shape[0]

#plt.scatter(X,Y)
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=1/10,random_state=0)

regressor = LinearRegression()
regressor.fit(X_train,Y_train)

plt.scatter(X_train,Y_train,color='blue')
plt.scatter(X_test,Y_test,color='red')

#plt.scatter(X,Y,color='black')
#plt.plot(X, regressor.predict(X), color='red')
```

```python
plt.scatter(X_train,Y_train,color='black')
plt.plot(X_train, regressor.predict(X_train), color='red')
```

```python


plt.scatter(X_test,Y_test,color='black')
plt.plot(X_test, regressor.predict(X_test), color='red')
```

Через псевдообратную матрицу

```python
N=int(n)
print(N)
K=2
F=np.zeros((N,K+1))

for i in range(N):
  F[i,0]=1

for i in range(N-1):
  for j in range(1,K+1):
    F[i,j]=XX[i,j-1]

TrF=np.transpose(F)
M=TrF.dot(F)
print(M)
Yt=np.transpose(Y)
rp=TrF.dot(Y)
print(np.transpose(rp))
M_inv = np.linalg.inv(M)
theta = M_inv.dot(rp)
print(theta)

YY=regressor2.predict([XX[134]])
print(YY)
Res=theta[0]+theta[1]*XX[134,0]+theta[2]*XX[134,1]
print(Res)

print(np.linalg.cond(M))
```

Нарисуем красиво плоскость

```python
x1min=min(x1)
x1max=max(x1)
print(x1max)
x2min=min(x2)
x2max=max(x2)
print(x2max)

xx1, xx2 = np.mgrid[x1min:x1max, x2min:x2max]
print(xx1[3,4])
zz = theta[0]+theta[1]*xx1+theta[2]*xx2

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(xx1, xx2, zz, cmap='inferno')
ax.legend()
```
